```python
dados['Number_of_Vehicles'].unique()   # volta os valores únicos que aparecem na coluna 'Number_of_Vehicles'
```

```python
for i in dados['Hour_of_Accident']:
  if (i <= 3) &(i >= 0):
   dados['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"0-3"})
  if (i <= 6) & (i > 3 ):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"3-6"})
  if (i <= 9) & (i > 6):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"6-9"})
  if (i <= 12) & (i > 9):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"9-12"})
  if (i <= 15) & (i > 12):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"12-15"})
  if (i <= 18) & (i > 15):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"15-18"})
  if (i <= 21) & (i > 18):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"18-21"})
  if (i <= 24) & (i > 21):
    dados ['Hour_of_Accident'] =  dados['Hour_of_Accident'].replace({i:"21-24"})
```
